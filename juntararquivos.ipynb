```python
import os

def juntar_arquivos_py(diretorio, arquivo_saida):
    # Abrir o arquivo de saída no modo de escrita
    with open(arquivo_saida, 'w') as arquivo_txt:
        # Percorrer todos os arquivos no diretório especificado
        for pasta_raiz, _, arquivos in os.walk(diretorio):
            for nome_arquivo in arquivos:
                # Verificar se o arquivo tem extensão .py
                if nome_arquivo.endswith('.py'):
                    caminho_arquivo = os.path.join(pasta_raiz, nome_arquivo)
                    with open(caminho_arquivo, 'r') as arquivo_py:
                        # Escrever o conteúdo do arquivo .py no arquivo de saída
                        arquivo_txt.write(f'# Arquivo: {nome_arquivo}\n')
                        arquivo_txt.write(arquivo_py.read())
                        arquivo_txt.write('\n\n')

# Defina o diretório onde estão os arquivos .py e o nome do arquivo de saída .txt
diretorio = r'D:\Nuvem\ghdaru\OneDrive\010_PROJETOS\3100_SIMULACAO\010_SIMULADORPYTHON\Inventory_Simulation\base'
arquivo_saida = 'saida.txt'

# Chamar a função para juntar os arquivos
juntar_arquivos_py(diretorio, arquivo_saida)
```
